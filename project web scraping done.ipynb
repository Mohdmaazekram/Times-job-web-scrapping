```python
Sta_urls = 'https://www.timesjobs.com/candidate/job-search.html?from=submit&searchType=Home_Search&luceneResultSize=25&postWeek=60&cboPresFuncArea=35&pDate=Y&sequence='
End_urls = '&startPage='
# This is Strart and End URLS.
```

```python
all_urls = []   # Taking a empty list.

for i in range(1,41):   # Using for loop to append all the URLS into empty list.
    all_urls.append(Sta_urls + str(i) + End_urls + str(i) )
#     print(all_urls)
```

```python
for item in all_urls:   # Taking all_urls from list and adjusting in sequence with help of for loop.
    print(item)     # Print to get out put.
```

```python
import requests    # Importing request Library.
from bs4 import BeautifulSoup    # Import BeautifulSoup from above.
```

```python
# def All_Pages_info():
    
all_page_lst = []  # Taking a empty list.

for url in all_urls:  # Using for loop to repeat all the urls line wise.

    all_url_data = requests.get(url).text  # Get the all_url_data from the URL website.

    all_soup_data = BeautifulSoup(all_url_data,'lxml') # Using BeautifulSoup to parse the all_url_data.

    all_lis = all_soup_data.find_all('li',class_='clearfix job-bx wht-shd-bx') # Creating one varialble to store common base of box.

    for item in all_lis:

        job_title = item.find('h2').text
        job_title1 = job_title.strip()
#             print(job_title1)                                            # Print to get job title1.

        company_name = item.find('h3',class_='joblist-comp-name').text
        company_name1 = company_name.replace('(More Jobs)','').strip()
#             print(company_name1)                                         # Print to get company_name1.

        experience = item.find('li').text
        experience1 = experience.strip('card_travel')
#             print(experience1)                                           # Print to get experience1.

        location = item.find('span').text.replace('(More Jobs)','')
#             print(location)                                              # Print to get location.

        job_description = item.find('ul',class_='list-job-dtl clearfix').text
        job_description1 = job_description.splitlines()[3:4]
        job_description2 = (','.join(job_description1))
#             print(job_description2)                                      # Print to get job_description2.

        key_skills = item.find('span',class_='srp-skills').text
        key_skills1 = key_skills.lstrip().splitlines()[:1]
        key_skills2 = (','.join(key_skills1))
#             print(key_skills2)                                            # Print to get key_skills2.

        job_detail_link = item.find('a')['href']
#             print(job_detail_link)                                       # Print to get job_detail_link.


        all_page_info = {                      # Store all keys and values in one dictionary.
            'Job_Title': job_title1,
            'Company_Name': company_name1,
            'Experience': experience1,
            'Location': location,
            'Job_Description': job_description2,
            'Key_Skills': key_skills2,
            'Job_Detail_Link': job_detail_link
        }

        all_page_lst.append(all_page_info)     # Taking all_page_info in empty list.
        print(all_page_lst)                                          # Print to get all_page_lst.
# All_Pages_info()
```

```python
import pandas as pd  # Import pandas Librarie.
```

```python
df = pd.DataFrame(all_page_lst)  # Createing DataFrame using panda.
```
